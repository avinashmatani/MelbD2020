```python
# Reading the 2019 dataset 

jan = pd.read_csv("DataSet/Jan19.csv")
janlast = jan.tail(1)
jan = jan[:-1]

feb = pd.read_csv("DataSet/Feb19.csv")
feblast = feb.tail(1)
feb = feb[:-1]
feb.append(janlast, ignore_index = True) 


mar = pd.read_csv("DataSet/Mar19.csv")
marlast = mar.tail(1)
mar = mar[:-1]
mar.append(feblast, ignore_index = True) 



apr = pd.read_csv("DataSet/Apr19.csv")
aprlast = apr.tail(1)
apr = apr[:-1]
apr.append(marlast, ignore_index = True) 


may = pd.read_csv("DataSet/May19.csv")
maylast = may.tail(1)
may = may[:-1]
may.append(aprlast, ignore_index = True) 



jun = pd.read_csv("DataSet/Jun19.csv")
junlast = jun.tail(1)
jun = jun[:-1]
jun.append(maylast, ignore_index = True) 



jul = pd.read_csv("DataSet/Jul19.csv")
jullast = jul.tail(1)
jul = jul[:-1]
jul.append(junlast, ignore_index = True) 



aug = pd.read_csv("DataSet/Aug19.csv")
auglast = aug.tail(1)
aug = aug[:-1]
aug.append(jullast, ignore_index = True) 



sep = pd.read_csv("DataSet/Sep19.csv")
seplast = sep.tail(1)
sep = sep[:-1]
sep.append(auglast, ignore_index = True) 


oct = pd.read_csv("DataSet/Oct19.csv")
octlast = oct.tail(1)
oct = oct[:-1]
oct.append(auglast, ignore_index = True) 



nov = pd.read_csv("DataSet/Nov19.csv")
novlast = nov.tail(1)
nov = nov[:-1]
nov.append(seplast, ignore_index = True) 


dec = pd.read_csv("DataSet/Dec19.csv")
declast = dec.tail(1)
dec = dec[:-1]
dec.append(novlast, ignore_index = True) 
```
